```python
class PrototypeScratchLinearRegression():
    """
    線形回帰のスクラッチ実装
    
    Parameters
    ----------
    num_iter : int
      イテレーション数
    lr : float
      学習率
    no_bias : bool
      バイアス項を入れない場合はTrue
    verbose : bool
      学習過程を出力する場合はTrue
    
    Attributes
    ----------
    self.coef_ : 次の形のndarray, shape (n_features,)
      パラメータ
    self.loss : 次の形のndarray, shape (self.iter,)
      訓練データに対する損失の記録
    self.val_loss : 次の形のndarray, shape (self.iter,)
      検証データに対する損失の記録
    """
    
    def __init__(self, num_iter, lr, no_bias, verbose=False):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.no_bias = no_bias
        self.verbose = verbose
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        # パラメータ
        self.coef_ = np.array([])
    def fit(self, X, y, X_val=None, y_val=None):
        """
        線形回帰を学習する。検証データが入力された場合はそれに対する損失と精度もイテレーションごとに計算する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        """
        if self.verbose:
            #verboseをTrueにした際は学習過程を出力
            print()

        # 特徴量の数（=Xの列数）に合わせてパラメータΘを初期化
        #self.theta = np.zeros(X.shape[1]).reshape(X.shape[1], 1)   #正式版
        self.theta = np.arange(1, X.shape[1] + 1).reshape(X.shape[1], 1)  #プロトタイプ用にダミーの値をセット（問題1のみで使用）

    def predict(self, X):
        """
        線形回帰を使い推定する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル
        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            線形回帰による推定結果
        """
        pass
        return

    def _linear_hypothesis(self, X):
        """
        線形の仮定関数を計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
          訓練データ
        Returns
        -------
        ret_y: 次の形のndarray, shape (n_samples, 1)
          線形の仮定関数による推定結果
        """
        
        # 特徴量の数: n_features (X.shape[1])
        # 線形重回帰の式: y = w0X0 + w1X1 + … + wnXn = x・w (x: n_samples x n_featuresの行列,  w(=theta): n_features x 1のベクトル)
        ret_y = X @ self.theta

        return ret_y
```

### ●ScratchLinearRegressionクラスのプロトタイプ2

```python
class PrototypeScratchLinearRegression2():
    """
    線形回帰のスクラッチ実装
    
    Parameters
    ----------
    num_iter : int
      イテレーション数
    lr : float
      学習率
    no_bias : bool
      バイアス項を入れない場合はTrue
    verbose : bool
      学習過程を出力する場合はTrue
    
    Attributes
    ----------
    self.coef_ : 次の形のndarray, shape (n_features,)
      パラメータ
    self.loss : 次の形のndarray, shape (self.iter,)
      訓練データに対する損失の記録
    self.val_loss : 次の形のndarray, shape (self.iter,)
      検証データに対する損失の記録
    """
    
    def __init__(self, num_iter, lr, no_bias, verbose=False):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.no_bias = no_bias
        self.verbose = verbose
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        # パラメータを空配列で初期化
        self.coef_ = np.array([])

        # デバッグ用情報出力
        if self.verbose:
            print("        イテレーション数:{}, 学習率:{}".format(self.iter, self.lr))

    def fit(self, X, y, X_val=None, y_val=None):
        """
        線形回帰を学習する。検証データが入力された場合はそれに対する損失と精度もイテレーションごとに計算する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        """
        # 特徴量の数（=Xの列数）に合わせてパラメータを初期化
        self.coef_ = np.zeros(X.shape[1])

        # デバッグ用情報出力
        if self.verbose:
            print("        パラメータ初期値:{}".format(self.coef_))

        # 最急降下法により最適なパラメータ（self.coef_）を決定
        self._gradient_descent(X, y)
        
    def predict(self, X):
        """
        線形回帰を使い推定する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル
        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            線形回帰による推定結果
        """
        pass
        return

    def _linear_hypothesis(self, X):
        """
        線形の仮定関数を計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        Returns
        -------
        ret_y: 次の形のndarray, shape (n_samples, 1)
            仮定関数による推定結果
        """
        
        # 特徴量の数: n_features (X.shape[1])
        # 線形重回帰の式: y = w0X0 + w1X1 + … + wnXn = x・w (x: n_samples x n_featuresの行列,  w(=coef_): n_features x 1のベクトル)
        ret_y_ndarray = X @ self.coef_.reshape(X.shape[1], 1)

        return ret_y_ndarray

    def _gradient_descent(self, X, y, error=0):
        """
        最急降下法によりパラメータを計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        error : ?
        Returns
        -------
        None
        """
        n_samples = X.shape[0]    # 全データ数
        n_features = X.shape[1]    # 特徴量の数
        batch_size = int(n_samples / (self.iter - 1))    #バッチサイズ（1回のパラメータ更新（=イテレーション）で使用するデータ数）
        
        update_counter = 0  # パラメータの更新回数

        # デバッグ用情報出力
        if self.verbose:
            print("        全データ数:{}, バッチサイズ:{}".format(n_samples, batch_size))

        #イテレーション数だけパラメータを更新する
        for k in range(self.iter):
            # デバッグ用情報出力
            if self.verbose:
                print("            イテレーション:{}回目".format(k))
            # バッチデータのインデックスが全データ数を超過しないようにバッチサイズを調整
            if(batch_size * (k + 1) > n_samples):
                batch_size = n_samples - (batch_size * k)
            # バッチデータの始点と終点を決定
            batch_start_index = k * batch_size
            batch_end_index = batch_start_index + batch_size
            # 学習データをバッチサイズ分だけ抽出
            X_devided = X[batch_start_index : batch_end_index, :]
            y_devided = y[batch_start_index : batch_end_index]
            # バッチデータから目的変数y（batch_size x 1のベクトル）を推定
            temp_y_pred_ndarray = self._linear_hypothesis(X_devided)
            # デバッグ用情報出力
            if self.verbose:
                # print("            X_devided:{}".format(X_devided))
                # print("            temp_y_pred:{}".format(temp_y_pred_ndarray))
                pass
            # 各特徴量のパラメータを順番に更新
            for j in range(n_features):
                # 勾配の値を初期化
                temp_gradient = 0
                # 現在のパラメータ値における勾配（Σ(h(xi)-yi)Xij）を計算
                for i in range(batch_size):
                    temp_gradient += (temp_y_pred_ndarray[i] - y_devided[i]) * X_devided[i, j]
                # パラメータを更新
                self.coef_[j] = self.coef_[j] - self.lr * (1/batch_size) * temp_gradient

                # デバッグ用情報出力
                if self.verbose:
                    print("                特徴量:{}番目".format(j))
                    print("                勾配:{}、パラメータ:{}".format(temp_gradient, self.coef_[j]))

            update_counter += 1
            # デバッグ用情報出力
            if self.verbose:
                print("                パラメータの更新回数:{}".format(update_counter))
        return
```

```python
class PrototypeScratchLinearRegression3():
    """
    線形回帰のスクラッチ実装
    
    Parameters
    ----------
    num_iter : int
      イテレーション数
    lr : float
      学習率
    no_bias : bool
      バイアス項を入れない場合はTrue
    verbose : bool
      学習過程を出力する場合はTrue
    
    Attributes
    ----------
    self.coef_ : 次の形のndarray, shape (n_features + 1,)   # [補足]特徴量の数に1を足しているのは、切片の分
      パラメータ
    self.intercept_ : int
      切片(=self.coef_[-1])
    self.loss : 次の形のndarray, shape (self.iter,)
      訓練データに対する損失の記録
    self.val_loss : 次の形のndarray, shape (self.iter,)
      検証データに対する損失の記録
    """
    
    def __init__(self, num_iter, lr, no_bias=True, verbose=False):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.no_bias = no_bias
        self.verbose = verbose
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        # パラメータを空配列で初期化
        self.coef_ = np.array([])   # 傾き（各特徴量に対する重み）
        self.intercept_ = 0   # 切片

        # デバッグ用情報出力
        if self.verbose:
            print("        イテレーション数:{}, 学習率:{}".format(self.iter, self.lr))

    def fit(self, X, y, X_val=None, y_val=None):
        """
        線形回帰を学習する。検証データが入力された場合はそれに対する損失と精度もイテレーションごとに計算する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        """
        # 切片への乗算用の列（全て1）を、Xに追加
        X = self._add_ones_vector(X)

        # 特徴量の数（=Xの列数）に合わせてパラメータを初期化
        self.coef_ = np.zeros(X.shape[1])

        # デバッグ用情報出力
        if self.verbose:
            print("        パラメータ初期値:{}".format(self.coef_))

        # 最急降下法により最適なパラメータ（self.coef_）を決定
        self._gradient_descent(X, y)

        # 切片に値をセット(self.coef_の最後が切片)
        self.intercept_ = self.coef_[-1]
        
    def predict(self, X):
        """
        線形回帰を使い推定する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            特徴量データ
        Returns
        -------
        ret_y_pred_ndarray : 次の形のndarray, shape (n_samples, 1)
            特徴量データに対する推定結果
              
        """
        # 切片への乗算用の列（全て1）を、Xに追加
        X = self._add_ones_vector(X)

        # データセットXから目的変数y（batch_size x 1のベクトル）を推定
        ret_y_pred_ndarray = self._linear_hypothesis(X)
        return ret_y_pred_ndarray

    def _linear_hypothesis(self, X):
        """
        線形の仮定関数を計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            特徴量
        Returns
        -------
        ret_y: 次の形のndarray, shape (n_samples, 1)
            仮定関数による推定結果
        """
        
        # 特徴量の数: n_features (X.shape[1])
        # 線形重回帰の式: y = w0X0 + w1X1 + … + wnXn = x・w (x: n_samples x n_featuresの行列,  w(=coef_): n_features x 1のベクトル)
        ret_y_ndarray = X @ self.coef_.reshape(X.shape[1], 1)

        return ret_y_ndarray

    def _gradient_descent(self, X, y, error=0):
        """
        最急降下法によりパラメータを計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        error : ?
        Returns
        -------
        None
        """
        n_samples = X.shape[0]    # 全データ数
        n_features = X.shape[1]    # 特徴量の数
        batch_size = int(n_samples / (self.iter - 1))    #バッチサイズ（1回のパラメータ更新（=イテレーション）で使用するデータ数）
        
        update_counter = 0  # パラメータの更新回数

        # デバッグ用情報出力
        if self.verbose:
            print("        全データ数:{}, バッチサイズ:{}".format(n_samples, batch_size))

        #イテレーション数だけパラメータを更新する
        for k in range(self.iter):
            # デバッグ用情報出力
            if self.verbose:
                print("            イテレーション:{}回目".format(k))
            # バッチデータのインデックスが全データ数を超過しないようにバッチサイズを調整
            if(batch_size * (k + 1) > n_samples):
                batch_size = n_samples - (batch_size * k)
            # バッチデータの始点と終点を決定
            batch_start_index = k * batch_size
            batch_end_index = batch_start_index + batch_size
            # 学習データをバッチサイズ分だけ抽出
            X_devided = X[batch_start_index : batch_end_index, :]
            y_devided = y[batch_start_index : batch_end_index]

            # バッチデータから目的変数y（batch_size x 1のベクトル）を推定
            temp_y_pred_ndarray = self._linear_hypothesis(X_devided)
            # デバッグ用情報出力
            if self.verbose:
                # print("            X_devided:{}".format(X_devided))
                # print("            temp_y_pred:{}".format(temp_y_pred_ndarray))
                pass
            # 各特徴量のパラメータを順番に更新
            for j in range(n_features):
                # 勾配の値を初期化
                temp_gradient = 0
                # 現在のパラメータ値における勾配（Σ(h(xi)-yi)Xij）を計算
                for i in range(batch_size):
                    temp_gradient += (temp_y_pred_ndarray[i] - y_devided[i]) * X_devided[i, j]
                # パラメータを更新
                self.coef_[j] = self.coef_[j] - self.lr * (1/batch_size) * temp_gradient

                # デバッグ用情報出力
                if self.verbose:
                    if (j == (n_features - 1)):
                        print("                切片:")
                    else:
                        print("                {}番目の特徴量:".format(j))
                    print("                    勾配:{}、パラメータ:{}".format(temp_gradient, self.coef_[j]))

            update_counter += 1
            # デバッグ用情報出力
            if self.verbose:
                print("                パラメータの更新回数:{}".format(update_counter))
        return

    def _add_ones_vector(self, X):
        """
        切片(=self.coef_[-1])への乗算用の列（全て1）を、Xに追加する関数
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        Returns
        -------
        None
        """
        ones_ndarray = np.ones((X.shape[0], 1))
        X = np.concatenate([X, ones_ndarray], axis=1)
        return X
```

```python
def MSE(y_true, y_pred):
    """
    平均二乗誤差の計算
    Parameters
    ----------
    y_pred : 次の形のndarray, shape (n_samples,)
      推定した値
    y_true : 次の形のndarray, shape (n_samples,)
      正解値
    Returns
    ----------
    mse : numpy.float
      平均二乗誤差
    """
    n_sample = len(y_pred) 
    # reshapeで列情報を明示的に付与
    y_true = y_true.reshape(y_true.shape[0], 1)
    y_pred = y_pred.reshape(y_pred.shape[0], 1)
    mse = (1 / n_sample) * ((y_pred - y_true)**2).sum()

    return mse
```

```python
class ScratchLinearRegression():
    """
    線形回帰のスクラッチ実装
    
    Parameters
    ----------
    num_iter : int
      イテレーション数
    lr : float
      学習率
    no_bias : bool
      バイアス項を入れない場合はTrue
    verbose : bool
      学習過程を出力する場合はTrue
    squared : bool
      lossの種類を選択する変数(True:MSE, False:RMSE)
    
    Attributes
    ----------
    self.coef_ : 次の形のndarray, shape (n_features + 1,)   # [補足]特徴量の数に1を足しているのは、切片の分
      パラメータ
    self.intercept_ : float
      切片(=self.coef_[-1])
    self.coef_log : 次の形のndarray, shape (self.iter, n_features + 1)   # [補足]特徴量の数に1を足しているのは、切片の分
      学習中のパラメータの更新ログ
    self.loss : 次の形のndarray, shape (self.iter,)
      訓練データに対する損失の記録
    self.val_loss : 次の形のndarray, shape (self.iter,)
      検証データに対する損失の記録
    """
    
    def __init__(self, num_iter, lr, no_bias=True, verbose=False, squared=True):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.no_bias = no_bias
        self.verbose = verbose
        self.squared = squared
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        # パラメータを空配列で初期化
        self.coef_ = np.array([])   # 傾き（各特徴量に対する重み）
        self.intercept_ = 0   # 切片
        self.coef_log = np.array([])   #パラメータ更新ログ

        # デバッグ用情報出力
        if self.verbose:
            print("        イテレーション数:{}, 学習率:{}".format(self.iter, self.lr))

    def fit(self, X, y, X_val=None, y_val=None):
        """
        線形回帰を学習する。検証データが入力された場合はそれに対する損失と精度もイテレーションごとに計算する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        """
        # 切片への乗算用の列（全て1）を、Xに追加
        X = self._add_ones_vector(X)

        # 特徴量の数（=Xの列数）に合わせてパラメータを初期化
        self.coef_ = np.zeros(X.shape[1])

        # パラメータ更新ログを初期化 (各列が特徴量ごとのパラメータとなるようにreshape)
        self.coef_log = self.coef_.reshape(1, self.coef_.shape[0]).copy()

        # デバッグ用情報出力
        if self.verbose:
            print("        パラメータ初期値:{}".format(self.coef_))
            print("        パラメータ更新ログ:{}".format(self.coef_log))

        # 最急降下法により最適なパラメータ（self.coef_）を決定
        self._gradient_descent(X, y)
        
        # 切片に値をセット(self.coef_の最後が切片)
        self.intercept_ = self.coef_[-1]

        return
        
    def predict(self, X):
        """
        線形回帰を使い推定する。
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            特徴量データ
        Returns
        -------
        ret_y_pred : 次の形のndarray, shape (n_samples, 1)
            特徴量データに対する推定結果
        """
        # 切片への乗算用の列（全て1）を、Xに追加
        X = self._add_ones_vector(X)

        # データセットXから目的変数y（n_sample x 1のベクトル）を推定
        ret_y_pred = self._linear_hypothesis(X, coef=self.coef_)

        return ret_y_pred

    def plot_learning_curve(self, X_test, y_test):
        """
        学習曲線をプロットする関数
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, 1)
            特徴量の検証データ
        y : 次の形のndarray, shape (n_samples, 1)
            目的変数の検証データ
        Returns
        -------
        None
        """
        # 引数をローカル変数に変換
        X = X_test
        y = y_test

        # 切片への乗算用の列（全て1）を、Xに追加
        X = self._add_ones_vector(X)

        # 検証データを対象にして損失関数を計算し、内部変数に記録
        self._calcurate_loss_to_val(X, y)

        # 学習データと検証データのそれぞれの損失関数の値をグラフに描画
        fig, ax = plt.subplots(figsize=(15, 10), dpi=50)
        ax.set_title("Learning Curve")
        ax.set_xlabel("Iteration Number")
        ax.set_ylabel("Loss")
        ax.plot(self.loss, color = "blue", label="train")
        ax.plot(self.val_loss, color = "red", label="validation")
        ax.legend(loc='best') # 凡例を最適位置に表示
        plt.show()

        return

    def _linear_hypothesis(self, X, coef):
        """
        線形の仮定関数を計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            特徴量
        Returns
        -------
        ret_y: 次の形のndarray, shape (n_samples, 1)
            仮定関数による推定結果
        """
        
        # 特徴量の数: n_features (X.shape[1])
        # 線形重回帰の式: y = w0X0 + w1X1 + … + wnXn = x・w (x: n_samples x n_featuresの行列,  w(=coef_): n_features x 1のベクトル)
        ret_y_ndarray = X @ coef.reshape(X.shape[1], 1)
        return ret_y_ndarray

    def _gradient_descent(self, X, y, error=0):
        """
        最急降下法によりパラメータを計算する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        error : ?
        Returns
        -------
        None
        """
        n_samples = X.shape[0]    # 全データ数
        n_features = X.shape[1]    # 特徴量の数
        update_counter = 0  # パラメータの更新回数
        
        # デバッグ用情報出力
        if self.verbose:
            print("        全データ数:{}:{}".format(n_samples))

        #イテレーション数だけパラメータを更新する
        for k in range(self.iter):
            # デバッグ用情報出力
            if self.verbose:
                print("            イテレーション:{}回目".format(k))

            # 特徴量データXから目的変数y（n_sample x 1のベクトル）を推定
            y_pred = self._linear_hypothesis(X, coef=self.coef_)

            # 各特徴量のパラメータを順番に更新
            for j in range(n_features):
                # 勾配の値を初期化
                temp_gradient = 0
                # 現在のパラメータ値における勾配（Σ(h(xi)-yi)Xij）を計算
                for i in range(n_samples):
                    temp_gradient += (y_pred[i] - y[i]) * X[i, j]
                # パラメータを更新
                self.coef_[j] = self.coef_[j] - self.lr * (1/n_samples) * temp_gradient

                # デバッグ用情報出力
                if self.verbose:
                    if (j == (n_features - 1)):
                        print("                切片:")
                    else:
                        print("                {}番目の特徴量:".format(j))
                    print("                    勾配:{}、パラメータ:{}".format(temp_gradient, self.coef_[j]))

            update_counter += 1

            # 損失関数を計算して内部変数に記録
            self.loss[k] = self._calcurate_loss(y, y_pred)
            # 現在のパラメータを更新ログに記録
            self.coef_log = np.append(self.coef_log, self.coef_.reshape(1, self.coef_.shape[0]), axis=0) 

            # デバッグ用情報出力
            if self.verbose:
                print("            パラメータの更新回数:{}".format(update_counter))
                # print("            パラメータ更新ログ:{}".format(self.coef_log))
        return

    def _add_ones_vector(self, X):
        """
        切片(=self.coef_[-1])への乗算用の列（全て1）を、Xに追加する関数
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        Returns
        -------
        ret_X : 次の形のndarray, shape (n_samples, n_features + 1)
            訓練データに、値が全て1の列を追加したもの
        """
        ones_ndarray = np.ones((X.shape[0], 1))
        ret_X = np.concatenate([X, ones_ndarray], axis=1)
        return ret_X

    def _calcurate_loss(self, y_true, y_pred):
        """
        損失関数の計算
        Parameters
        ----------
        y_pred : 次の形のndarray, shape (n_samples,)
          推定した値
        y_true : 次の形のndarray, shape (n_samples,)
          正解値
        Returns
        ----------
        mse : numpy.float
          平均二乗誤差
        """
        n_sample = len(y_pred)

        # yの推定値と正解値の配列に、明示的に列情報を付与
        y_pred = y_pred.reshape(y_pred.shape[0], 1)
        y_true = y_true.reshape(y_true.shape[0], 1)

        loss = (1 / (2 * n_sample)) * (((y_pred - y_true)**2).sum())
        # loss = 0
        # for i, _ in enumerate(y_pred):
        #     loss += (y_pred[i] - y_true[i])**2
        #     print("                loss:\n{}".format(loss))

        # squared=Falseの時、lossをMSE→RMSEに変換
        if (self.squared == False):
            loss = loss**(1/2)

        # デバッグ用情報出力
        if self.verbose:
            # print("                y_pred:\n{}".format(y_pred))
            # print("                y_true:\n{}".format(y_true))
            print("                loss:{}".format(loss))
        return loss

    def _calcurate_loss_to_val(self, X_test, y_test):
        """
        検証データに対する損失を算出する関数
        Parameters
        ----------
        X_test : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_test : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        Returns
        -------
        None
        """
        n_samples = X.shape[0]    # 検証データの数
        n_features = X.shape[1]    # 特徴量の数
        
        # デバッグ用情報出力
        if self.verbose:
            print("            X_test:{}".format(X_test))
            print("            y_test:{}".format(y_test))

        #イテレーション数だけ検証データに対する損失関数を算出する
        for k in range(self.iter):
            # デバッグ用情報出力
            if self.verbose:
                print("            イテレーション:{}回目".format(k))

            # 更新ログ内のパラメータを先頭から順に用いて、検証データから目的変数y（n_samples x 1のベクトル）を推定
            y_pred = self._linear_hypothesis(X_test, coef=self.coef_log[k])

            # デバッグ用情報出力
            if self.verbose:
                print("            y_pred:{}".format(y_pred))
            
            # 損失関数を計算して内部変数に記録
            self.val_loss[k] = self._calcurate_loss(y_test, y_pred)
        return
```
